```python
df = pd.read_csv("cleaned_orders.csv")

print("Loaded Dataset")
df.head()
```

```python
df.loc[
    (df["expected_date"] <= df["actual_date"]) &
    (df["delivery_status"] == "Pending"),
    "delivery_status"
] = "Delivered"

print("Updated Delivery Status")
df.head()
```

```python
df["expected_date"] = pd.to_datetime(df["expected_date"])
df["actual_date"] = pd.to_datetime(df["actual_date"])

df["delay_days"] = (
    df["actual_date"] - df["expected_date"]
).dt.days

df.head()
```

```python
df["is_delayed"] = df["delay_days"].apply(
    lambda x: "Yes" if x > 0 else "No"
)

df.head()
```

```python
df.to_csv("final_orders_output.csv", index=False)

print("Final Output Saved")
```
