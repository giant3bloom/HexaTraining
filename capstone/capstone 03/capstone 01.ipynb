```python
import csv
from pandas import read_csv
import numpy as np
file_path = r'C:\hex-t\HexaTraining\capstone\capstone 03\dataset\orders.csv'

with open(file_path, 'r') as f:
    read = csv.reader(f)
    rows = list(read)
    for i in rows:
        print(i)
    count = len(rows) - 1
    print('total count: ', count)

    revenue = 0
    for row in rows[1:]:
        try:
            revenue += (int(row[5]) * int(row[6]))
        except:
            continue
    print('total revenue: ', revenue)

    highest_order_value = 0
    for row in rows[1:]:
        try:
            highest_order_value = max(
                highest_order_value,
                (int(row[5]) * int(row[6]))
            )
        except:
            continue
    print('max val: ', highest_order_value)

    lowest_order_value = float('inf')
    for row in rows[1:]:
        try:
            lowest_order_value = min(
                lowest_order_value,
                (int(row[5]) * int(row[6]))
            )
        except:
            continue
    print('min val: ', lowest_order_value)

    lst = []
    for row in rows[1:]:
        try:
            lst.append(int(row[5]) * int(row[6]))
        except:
            continue
    print('avg_val: ', sum(lst)/len(lst))
```
