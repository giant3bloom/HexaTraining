```python
def find_top_scorer():
    return df[df['runs'] == df['runs'].max()]

def calculate_average_runs():
    return df['runs'].mean()

def find_best_team():
    team_runs = df.groupby('team')['runs'].sum()
    best_team = team_runs.idxmax()
    return best_team, team_runs.max()

def find_total_boundaries():
    return {
        "total_fours": df['fours'].sum(),
        "total_sixes": df['sixes'].sum()
    }
```

```python
print('top score')
print(find_top_scorer())
print('avg run')
print(calculate_average_runs())
print('best team')
print(find_best_team())
print('total boundaries')
print(find_total_boundaries())
```

```python
import pandas as pd

df = pd.read_csv(file_path)
total_players = len(df)
total_runs = df['runs'].sum()
avg_runs = df['runs'].mean()
high_scorer = df.loc[df['runs'].idxmax(), 'player_name']
high_runs = df['runs'].max()
low_scorer = df.loc[df['runs'].idxmin(), 'player_name']
low_runs = df['runs'].min()
team_runs = df.groupby('team')['runs'].sum()
top5_players = df.nlargest(5, 'runs')[['player_name', 'runs']]
most_4_player = df.loc[df['fours'].idxmax(), 'player_name']
most_4 = df['fours'].max()
most_6_player = df.loc[df['sixes'].idxmax(), 'player_name']
most_6 = df['sixes'].max()

w_file = r'C:\hex-t\HexaTraining\capstone\capstone 03\generated_dataset\cricket_report.txt'

with open(w_file, 'w') as w_f:
    w_f.write("CRICKET REPORT\n")
    w_f.write("====================\n\n")
    w_f.write(f"Total Players : {total_players}\n")
    w_f.write(f"Total Runs : {total_runs}\n")
    w_f.write(f"Average Runs : {avg_runs:.2f}\n")
    w_f.write(f"Highest Scorer : {high_scorer} ({high_runs})\n")
    w_f.write(f"Lowest Scorer : {low_scorer} ({low_runs})\n\n")
    w_f.write("Team Wise Runs:\n")
    for team, runs in team_runs.items():
        w_f.write(f"{team} : {runs}\n")
    w_f.write("\nTop 5 Players:\n")
    for _, row in top5_players.iterrows():
        w_f.write(f"{row['player_name']} : {row['runs']}\n")
    w_f.write(f"\nMost Fours : {most_4_player} ({most_4})\n")
    w_f.write(f"Most Sixes : {most_6_player} ({most_6})\n")

print("Report generated successfully")
```

```python
import pandas as pd

file_path = r'C:\hex-t\HexaTraining\capstone\capstone 03\dataset\players.csv'

df = pd.read_csv(file_path)

def player_analysis():
    print("\nPLAYER ANALYSIS")
    print("===================")
    print("\nTop Scorer:")
    print(df[df['runs'] == df['runs'].max()])
    print("\nAverage Runs:")
    print(df['runs'].mean())
    print("\nTop 5 Players:")
    print(df.nlargest(5, 'runs')[['player_name', 'runs']])


def team_analysis():
    print("\nTEAM ANALYSIS")
    print("===================")
    print("\nTeam Wise Runs:")
    print(df.groupby('team')['runs'].sum())
    print("\nAverage Runs By Team:")
    print(df.groupby('team')['runs'].mean())
    print("\nTop Team:")
    print(df.groupby('team')['runs'].sum().nlargest(1))


def boundary_analysis():
    print("\nBOUNDARY ANALYSIS")
    print("===================")
    print("\nMost Fours:")
    print(df.loc[df['fours'].idxmax(), ['player_name', 'fours']])
    print("\nMost Sixes:")
    print(df.loc[df['sixes'].idxmax(), ['player_name', 'sixes']])
    print("\nTotal Fours:")
    print(df['fours'].sum())
    print("\nTotal Sixes:")
    print(df['sixes'].sum())


def export_reports():
    report_file = r'C:\hex-t\HexaTraining\capstone\capstone 03\generated_dataset\cricket_report.txt'
    with open(report_file, 'w') as w_f:
        w_f.write("CRICKET REPORT\n")
        w_f.write("====================\n\n")
        w_f.write(f"Total Players : {len(df)}\n")
        w_f.write(f"Total Runs : {df['runs'].sum()}\n")
        w_f.write(f"Average Runs : {df['runs'].mean():.2f}\n")
    print("\nReport Exported Successfully")


while True:

    print("\n====== CRICKET ANALYSIS SYSTEM ======")
    print("1. Player Analysis")
    print("2. Team Analysis")
    print("3. Boundary Analysis")
    print("4. Export Reports")
    print("5. Exit")

    choice = input("\nEnter your choice: ")

    if choice == '1': player_analysis()
    elif choice == '2': team_analysis()
    elif choice == '3': boundary_analysis()
    elif choice == '4': export_reports()
    elif choice == '5':
        print("Exiting Application...")
        break
    else:
        print("Invalid Choice")
```
