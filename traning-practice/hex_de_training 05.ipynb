```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('pyspark day 1').getOrCreate()
```

```python
data = [
    (101, "Rahul Sharma", "Data Engineering", 75000, "Hyderabad"),
    (102, "Priya Reddy", "AI Engineering", 85000, "Bangalore"),
    (103, "Amit Kumar", "Data Engineering", 65000, "Mumbai"),
    (104, "Sneha Patel", "Data Science", 95000, "Chennai"),
    (105, "Farhan Ali", "Cloud Engineering", 80000, "Delhi")
]

columns = [
    "employee_id",
    "name",
    "department",
    "salary",
    "city"
]
```

```python
df = spark.createDataFrame(data, columns)
df.show()
```

```python
df.printSchema()
```

```python
df.filter(df.salary > 80000).show()
```

```python
from pyspark.sql.functions import col

df_bonus = df.withColumn(
    'bonus',
    col('salary') * 0.10
)

df_bonus.show()
```

```python
df.filter(df.department.isin('Data Engineering', 'AI Engineering')).show()
df.orderBy('salary').show()
df.orderBy(df.salary.desc()).show()
df.orderBy('name').show()
df.orderBy('city', 'salary').show()
from pyspark.sql.functions import col
df_bonus = df.withColumn(
    'bonus',
    col('salary') * 0.10
)
df_bonus.show()
df_tax = df.withColumn(
    'tax',
    col('salary') * 0.05
)
df_tax.show()
```

```python
df_annual = df.withColumn(
    'annual_salary',
    col('salary') * 12
)
df_annual.show()

df.withColumn(
    'salary_after_bouns',
    col('salary') + col('salary') * 0.10
).show()

df_renamed = df.withColumnRenamed(
    'salary',
    'monthly_salary'
)
df_renamed.show()

df.withColumnRenamed(
    'department',
    'team'
).show()

from pyspark.sql.functions import min, max, sum, avg
print('total payout: ', df.select(sum('salary')).show())
print('avg salary: ', df.select(avg('salary')).show())
print('max salary: ', df.select(max('salary')).show())
print('min salary: ', df.select(min('salary')).show())
print('employ count: ', df.count())
df.groupby('department').count().show()

```

```python
from pyspark.sql.functions import col

df.filter(col('name').like('R%')).show()
df.filter(col('name').like('%a%')).show()
df.filter(col('department').contains('Engineering')).show()
```

```python
from pyspark.sql.functions import when, col

df_category = df.withColumn(
    'salary_category',
    when(col('salary') > 85000, 'High')
    .when(col('salary') > 75000, 'Medium')
    .otherwise('Low')
)

df_category.show()
```

```python
df_category.groupBy('salary_category').count().show()
```

```python
df_summary = df.withColumns({
    'bonus': col('salary') * 0.1,
    'annual salary': col('salary') * 12
})

df_summary.select(*[col for col in df_summary.columns if col != 'city']).show()
```
