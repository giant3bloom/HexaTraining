```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('RDD').getOrCreate()
```

```python
number = [10, 20, 30, 40, 50]
rdd = spark.sparkContext.parallelize(number)

print(rdd.collect())
```

```python
activities = [
  "LOGIN",
  "COURSE_VIEW",
  "LOGIN",
  "VIDEO_PLAY",
  "QUIZ_START",
  "VIDEO_PLAY",
  "LOGIN",
  "COURSE_VIEW",
  "ASSIGNMENT_SUBMIT",
  "VIDEO_PLAY",
  "QUIZ_START",
  "QUIZ_COMPLETE",
  "LOGIN",
  "VIDEO_PLAY",
  "COURSE_VIEW",
  "LOGIN",
  "VIDEO_PLAY",
  "VIDEO_PLAY",
  "QUIZ_START",
  "LOGIN",
  "COURSE_VIEW",
  "COURSE_VIEW",
  "ASSIGNMENT_SUBMIT",
  "LOGIN",
  "VIDEO_PLAY",
  "QUIZ_COMPLETE",
  "LOGIN",
  "COURSE_VIEW",
  "VIDEO_PLAY",
  "VIDEO_PLAY",
  "LOGIN",
  "COURSE_VIEW",
  "QUIZ_START",
  "ASSIGNMENT_SUBMIT",
  "LOGIN",
  "VIDEO_PLAY",
  "LOGIN",
  "COURSE_VIEW",
  "QUIZ_COMPLETE",
  "VIDEO_PLAY",
  "LOGIN",
  "COURSE_VIEW",
  "VIDEO_PLAY",
  "LOGIN",
  "QUIZ_START",
  "COURSE_VIEW",
  "VIDEO_PLAY",
  "ASSIGNMENT_SUBMIT",
  "LOGIN",
  "VIDEO_PLAY"
]

rdd = spark.sparkContext.parallelize(activities)
```

```python
rdd.collect()
print('rdd count: ', rdd.count())
print('rdd 1st activity: ', rdd.take(1))
print('rdd 1st -> 10th activity :')
print(rdd.take(10))
print(rdd.collect()[-1])
print('rdd unique : ', rdd.distinct().collect())
print('rdd unique count: ', rdd.distinct().count())
print('rdd sort: ', rdd.sortBy(lambda x: x).collect())
print('rdd upper: ', rdd.map(lambda x: x.upper()).collect())
print('rdd name : len')
for name in rdd.collect():
    print(name, len(name))
```

```python
print(rdd.filter(lambda x: x== 'LOGIN').collect())
print(rdd.filter(lambda x: x== 'VIDEO_PLAY').collect())
print(rdd.filter(lambda x: x== 'QUIZ_START').collect())
print(rdd.filter(lambda x: 'QUIZ' in x).collect())
print(rdd.filter(lambda x: x.endswith('VIEW')).collect())
print(rdd.filter(lambda x: x != 'LOGIN').collect())
print(rdd.filter(lambda x: x != 'VIDEO_PLAY').collect())
print(rdd.filter(lambda x: len(x) > 8).collect())
print(rdd.filter(lambda x: len(x) < 10).collect())
print(rdd.filter(lambda x: x.startswith('C')).collect())
```

```python
print(rdd.map(lambda x: x.lower()).collect())
print(rdd.map(lambda x: 'EVENT_' + x).collect())
print(rdd.map(lambda x: x + '_2025').collect())
print(rdd.map(lambda x: len(x)).collect())
print(rdd.map(lambda x: (x, len(x))).collect())
print(rdd.map(lambda x: (x, 1)).collect())
print(rdd.map(lambda x: 'USER_LOGIN' if x=='LOGIN' else x).collect())
print(rdd.map(lambda x: 'WATCH_VIDEO' if x== 'VIDEO_PLAY' else x).collect())
print(rdd.map(lambda x: x[::-1]).collect())
print(rdd.map(lambda x: x[0]).collect())
```

```python
print(rdd.distinct().collect())
print(rdd.distinct().count())
print(rdd.distinct().sortBy(lambda x: x).collect())
print(rdd.filter(lambda x: 'QUIZ' in x).distinct().collect())
print(rdd.filter(lambda x: 'COURSE' in x).distinct().collect())

activity_count = rdd.map(lambda x: (x, 1)).reduceByKey(lambda a, b: a + b)

print(activity_count.collect())
print(activity_count.take(1)[0])
print(activity_count.collect()[-1])
print(activity_count.sortBy(lambda x: x[1]).collect())
print(activity_count.filter(lambda x: x[1] > 5).collect())
```

```python
numbers = [
  120, 250, 300, 450, 600, 750, 900, 1050,
  1200, 1350, 1500, 1650, 1800, 1950, 2100
]

sales_rdd = spark.sparkContext.parallelize(numbers)
```

```python
print('total: ', sales_rdd.reduce(lambda a,b: a+b))
print('max : ', sales_rdd.reduce(lambda a,b: max(a,b)))
print('min : ', sales_rdd.reduce(lambda a,b: min(a,b)))
print('avg : ', sales_rdd.reduce(lambda a,b: a+b) / rdd.count())
print('>1k : ', sales_rdd.filter(lambda x: x>1000).count())
print('2x : ', sales_rdd.map(lambda x: x*2).collect())
print('+100 : ', sales_rdd.map(lambda x: x+100).collect())
print('sort asc: ', sales_rdd.sortBy(lambda x: x).collect())
print('sort desc: ', sales_rdd.sortBy(lambda x: -x).collect())
print('top 5: ', sales_rdd.sortBy(lambda x: -x).take(5))
```

```python
courses = [
  "Python SQL Spark",
  "Azure Databricks Spark",
  "Python Pandas NumPy",
  "SQL Spark DeltaLake",
  "Azure Python SQL"
]

course_rdd = spark.sparkContext.parallelize(courses)
```

```python
print(course_rdd.flatMap(lambda x: x.split()).collect())
print(course_rdd.flatMap(lambda x: x.split()).count())
print(course_rdd.flatMap(lambda x: x.split()).distinct().collect())
print(course_rdd.flatMap(lambda x: x.split()).sortBy(lambda x: x).collect())
print(course_rdd.flatMap(lambda x: x.split()) \
      .map(lambda x: (x, 1)).reduceByKey(lambda a,b : a+b).collect())
print(course_rdd.flatMap(lambda x: x.split()).filter(lambda x: 'Python' in x).collect())
print(course_rdd.flatMap(lambda x: x.split()).filter(lambda x: 'Spark' in x).collect())
print(course_rdd.flatMap(lambda x: x.split()).map(lambda x: x.upper()).collect())
print(course_rdd.flatMap(lambda x: x.split()).map(lambda x: len(x)).reduce(lambda a,b: max(a,b)))
print(course_rdd.flatMap(lambda x: x.split()).map(lambda x: len(x)).reduce(lambda a,b: min(a,b)))
```
