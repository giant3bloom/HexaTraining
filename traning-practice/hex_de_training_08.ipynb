```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('session 03').getOrCreate()
```

```python
transactions = [
  "DEPOSIT",
  "WITHDRAW",
  "TRANSFER",
  "DEPOSIT",
  "BALANCE_CHECK",
  "TRANSFER",
  "DEPOSIT",
  "WITHDRAW",
  "LOAN_PAYMENT",
  "DEPOSIT",
  "TRANSFER",
  "WITHDRAW",
  "DEPOSIT",
  "BALANCE_CHECK",
  "TRANSFER",
  "DEPOSIT",
  "WITHDRAW",
  "LOAN_PAYMENT",
  "TRANSFER",
  "DEPOSIT",
  "WITHDRAW",
  "DEPOSIT",
  "TRANSFER",
  "BALANCE_CHECK",
  "WITHDRAW",
  "DEPOSIT",
  "TRANSFER",
  "LOAN_PAYMENT",
  "DEPOSIT",
  "WITHDRAW",
  "TRANSFER",
  "DEPOSIT",
  "BALANCE_CHECK",
  "TRANSFER",
  "WITHDRAW",
  "DEPOSIT",
  "TRANSFER",
  "LOAN_PAYMENT",
  "DEPOSIT",
  "WITHDRAW",
  "TRANSFER",
  "BALANCE_CHECK",
  "DEPOSIT",
  "TRANSFER",
  "WITHDRAW",
  "LOAN_PAYMENT",
  "DEPOSIT",
  "TRANSFER",
  "BALANCE_CHECK",
  "DEPOSIT"
]

rdd = spark.sparkContext.parallelize(transactions)
```

```python
print(rdd.collect())
print(rdd.count())
print(rdd.take(1)[0])
print(rdd.take(15))
print(rdd.distinct().collect())
print(rdd.distinct().count())
print(rdd.sortBy(lambda x: x).collect())
print(rdd.map(lambda x: x.lower()).collect())
print(rdd.map(lambda x: x.upper()).collect())
print(rdd.distinct().map(lambda x: (x, len(x))).collect())
```

```python
print(rdd.filter(lambda x: x== 'DEPOSIT').collect())
print(rdd.filter(lambda x: x== 'WITHDRAW').collect())
print(rdd.filter(lambda x: x== 'TRANSFER').collect())
print(rdd.filter(lambda x: x== 'LOAN_PAYMENT').collect())
print(rdd.filter(lambda x: 'CHECK' in x).collect())
print(rdd.filter(lambda x: x != 'DEPOSIT').collect())
print(rdd.filter(lambda x: x != 'TRANSFER').collect())
print(rdd.filter(lambda x: len(x) > 8).collect())
print(rdd.filter(lambda x: len(x) < 10).collect())
print(rdd.filter(lambda x: x.startswith('B')).collect())
```

```python
print(rdd.map(lambda x: 'BANK_' + x).collect())
print(rdd.map(lambda x: x + '_2025').collect())
print(rdd.map(lambda x: (x, len(x))).collect())
print(rdd.map(lambda x: (x, 1)).collect())
print(rdd.map(lambda x: 'CASE_DEPOSIT' if x== 'DEPOSIT' else x).collect())
print(rdd.map(lambda x: 'CASE_WITHDRAWAL' if x== 'WITHDRAW' else x).collect())
print(rdd.map(lambda x: x[::-1]).collect())
print(rdd.map(lambda x: x[:3]).collect())
print(rdd.map(lambda x: x[-3::]).collect())
dit = {'DEPOSIT': 'D', 'WITHDRAW': 'W', 'TRANSFER': 'T'}
print(rdd.map(lambda x: dit.get(x, x)).collect())
```

```python
print(rdd.distinct().collect())
print(rdd.distinct().count())
print(rdd.distinct().sortBy(lambda x: x).collect())
print(rdd.filter(lambda x: 'PAYMENT' in x).distinct().collect())
print(rdd.filter(lambda x: 'CHECK' in x).distinct().collect())
print(rdd.map(lambda x: (x, 1)).reduceByKey(lambda a,b : a+b).collect())
print(rdd.map(lambda x: (x, 1)).reduceByKey(lambda a,b : a+b).sortBy(lambda x: x[1]).take(1)[0])
print(rdd.map(lambda x: (x, 1)).reduceByKey(lambda a,b : a+b).sortBy(lambda x: -x[1]).take(1)[0])
print(rdd.map(lambda x: (x, 1)).reduceByKey(lambda a,b : a+b).collect())
print(rdd.map(lambda x: (x, 1)).reduceByKey(lambda a,b : a+b).filter(lambda x: x[1]>5).collect())
```

```python
amounts = [
  500,1200,3000,4500,800,950, 1100,2500,4000,6000,
  750,1800,2200,3300,4700, 5200,1500,2600,3700,4800,
  5900,6200,700,900,1400, 1750,2100,2900,3600,4300,
  5100,5800,6500,7200,8100, 9200,10500,12000,13500,15000,
  17000,19000,22000,25000,28000, 30000,32000,35000,38000,40000
]
amount_rdd = spark.sparkContext.parallelize(amounts)
```

```python
print('sum : ', amount_rdd.reduce(lambda a,b : a+b))
print('avg : ', amount_rdd.reduce(lambda a,b : a+b) / amount_rdd.count())
print('max : ', amount_rdd.reduce(lambda a,b : max(a,b)))
print('min : ', amount_rdd.reduce(lambda a,b : min(a,b)))
print('>10000 : ', amount_rdd.filter(lambda x: x>10000).collect())
print('<2000 : ', amount_rdd.filter(lambda x: x<2000).collect())
print('+500 : ', amount_rdd.map(lambda x: x+500).collect())
print('110% : ', amount_rdd.map(lambda x: x*1.1).collect())
print('sort asc : ', amount_rdd.sortBy(lambda x: x).collect())
print('sort desc : ', amount_rdd.sortBy(lambda x: -x).collect())
print('top 10 : ', amount_rdd.sortBy(lambda x: -x).take(10))
print('bottom 10 : ', amount_rdd.sortBy(lambda x: x).take(10))
print('count >5000 : ', amount_rdd.filter(lambda x: x > 5000).count())
print('count >20000 : ', amount_rdd.filter(lambda x: x > 20000).count())
print('total >10000 : ', amount_rdd.filter(lambda x: x > 10000).reduce(lambda a,b : a+b))
```

```python
customer_comments = [
  "service was excellent",
  "mobile banking is easy",
  "transfer completed quickly",
  "deposit process was smooth",
  "excellent customer support",
  "loan payment completed successfully",
  "banking app is excellent",
  "service is fast and reliable"
]
comment_rdd = spark.sparkContext.parallelize(customer_comments)
```

```python
grand_rdd = rdd.union(rdd2).union(rdd3)

print(grand_rdd.count())
print(grand_rdd.map(lambda x: (x, 1)).reduceByKey(lambda a,b : a+b).sortBy(lambda x: -x[1]).take(3))
print(grand_rdd.map(lambda x: (x, 1)).reduceByKey(lambda a,b : a+b).sortBy(lambda x: x[1]).take(3))
print(grand_rdd.map(lambda x: (x, 1)).reduceByKey(lambda a,b : a+b).sortBy(lambda x: x[1]).collect())
```

```python
print('FINAL BANKING ANALYTICS REPORT')
print('==============================')

print('\nTotal Transactions : ')
print(grand_rdd.count())

print('\nUnique Transaction Types : ')
print(grand_rdd.distinct().collect())

print('\nTotal Unique Transaction Types : ')
print(grand_rdd.distinct().count())

print('\nTransaction Frequency Report : ')
freq_rdd = grand_rdd.map(lambda x: (x, 1)) \
                    .reduceByKey(lambda a, b: a + b)

for i in freq_rdd.collect():
    print(i[0], ' : ', i[1])

print('\nMost Frequent Transaction : ')
print(
    freq_rdd.sortBy(lambda x: -x[1])
            .take(1)[0]
)

print('\nLeast Frequent Transaction : ')
print(
    freq_rdd.sortBy(lambda x: x[1])
            .take(1)[0]
)

print('\nTransactions Containing PAYMENT : ')
print(
    grand_rdd.filter(lambda x: 'PAYMENT' in x)
             .distinct()
             .collect()
)

print('\nTransactions Starting With B : ')
print(
    grand_rdd.filter(lambda x: x.startswith('B'))
             .collect()
)

print('\nSorted Transactions : ')
print(
    grand_rdd.sortBy(lambda x: x)
             .collect()
)
```
