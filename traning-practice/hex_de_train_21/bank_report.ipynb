```python
bank_data = [
    (1001, "Rahul", "Savings", 50000, "Hyderabad"),
    (1002, "Priya", "Current", 75000, "Chennai"),
    (1003, "Arjun", "Savings", 120000, "Bangalore"),
    (1004, "Sneha", "Current", 90000, "Mumbai"),
    (1005, "Kiran", "Savings", 65000, "Delhi")
]

columns = [
    "account_id",
    "customer_name",
    "account_type",
    "balance",
    "city"
]

bank_df = spark.createDataFrame(bank_data, columns)

bank_df.show()
```

```python
from pyspark.sql.functions import sum

bank_df.groupBy('city').agg(sum('balance')).show()
```
