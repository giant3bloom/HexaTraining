```python
product_data = [
    (1001, "Laptop", "Electronics", 65000),
    (1002, "Smartphone", "Electronics", 30000),
    (1003, "Headphones", "Accessories", 2500),
    (1004, "Office Chair", "Furniture", 7000),
    (1005, "Water Bottle", "Home & Kitchen", 500)
]

columns = ["product_id", "product_name", "category", "price"]
```

```python
products_df = spark.createDataFrame(product_data, columns)
display(products_df)
```

```python
products_df.createOrReplaceTempView("products")
```

```python
%sql
select avg(price) from products;
```
