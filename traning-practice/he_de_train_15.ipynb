```python

df = spark.read.csv(
    "/Volumes/learn_databricks/default/my_volume/sales_csv",
    header=True,
    inferSchema=True
)


df = (
    df
    .withColumn("price", col("price").cast("double"))
    .withColumn("quantity", col("quantity").cast("int"))
    .withColumn("revenue", col("price") * col("quantity"))
    .withColumn("tax", col("price") * col("quantity") * 0.05)
    .withColumn("net_revenue", col("price") * col("quantity") * 1.05)
)


completed_df = df.filter(col("status") == "Completed")


completed_df.select(
    "order_id",
    "product",
    "revenue"
).show()


city_revenue = (
    completed_df
    .groupBy("city")
    .agg(sum("revenue").alias("city_revenue"))
)

city_revenue.show()


category_revenue = (
    completed_df
    .groupBy("category")
    .agg(sum("revenue").alias("category_revenue"))
)

category_revenue.show()


completed_df.write \
    .mode("overwrite") \
    .parquet("/Volumes/learn_databricks/default/my_volume/sales_parquet")


parquet_df = spark.read.parquet(
    "/Volumes/learn_databricks/default/my_volume/sales_parquet"
)


parquet_df.createOrReplaceTempView("sales_parquet")


spark.sql("""
SELECT
    city,
    category,
    COUNT(*) AS total_orders,
    SUM(revenue) AS total_revenue,
    AVG(revenue) AS avg_revenue,
    MAX(revenue) AS max_revenue,
    MIN(revenue) AS min_revenue
FROM sales_parquet
GROUP BY city, category
ORDER BY total_revenue DESC
""").show()
```
