```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, max, min, avg, sum
```

```python
data = [
    (1, "Rahul Sharma", "Hyderabad", "Electronics", "Laptop", 1, 65000, "Delivered"),
    (2, "Priya Reddy", "Bangalore", "Electronics", "Mobile", 2, 25000, "Delivered"),
    (3, "Amit Kumar", "Mumbai", "Furniture", "Chair", 4, 3500, "Pending"),
    (4, "Sneha Patel", "Chennai", "Electronics", "Tablet", 1, 30000, "Cancelled"),
    (5, "Farhan Ali", "Delhi", "Furniture", "Table", 2, 12000, "Delivered"),
    (6, "Neha Singh", "Hyderabad", "Fashion", "Shoes", 3, 2500, "Delivered"),
    (7, "Arjun Verma", "Pune", "Fashion", "Watch", 1, 8000, "Pending"),
    (8, "Meera Nair", "Bangalore", "Electronics", "TV", 1, 45000, "Delivered"),
    (9, "Kiran Rao", "Hyderabad", "Furniture", "Desk", 1, 15000, "Delivered"),
    (10, "Ravi Kumar", "Mumbai", "Electronics", "Mobile", 1, 28000, "Delivered")
]

columns = [
    "order_id",
    "customer_name",
    "city",
    "category",
    "product",
    "quantity",
    "price",
    "order_status"
]
```

```python
spark = SparkSession.builder.appName('session 02').getOrCreate()

df = spark.createDataFrame(data, columns)

df.show()
```

```python
df.show()
df.select('customer_name', 'city', 'product').show()
df.select('product', 'price').distinct().show()
df.filter(df.city == 'Hyderabad').show()
df.filter(df.city == 'Bangalore').show()
df.filter(df.category == 'Electronics').show()
df.filter(df.category == 'Furniture').show()
df.filter(df.price > 30000).show()
df.filter(df.quantity > 1).show()
df.filter(df.order_status == 'Delivered').show()
```

```python
df.filter(df.order_status == 'Pending').show()
df.filter(df.order_status == 'Cancelled').show()
df.filter(df.city.isin('Hyderabad', 'Mumbai')).show()
df.filter(df.category.isin('Electronics', 'Fashion')).show()
df.filter(df.price.between(10000, 50000)).show()
df.orderBy(df.price).show()
df.orderBy(df.price.desc()).show()
df.orderBy('city', 'price').show()

full_df = df.withColumn(
    'total_amount',
    col('quantity') * col('price')
)
full_df.show()

full_df = full_df.withColumn(
    'discount',
    col('total_amount') * 0.10
)
full_df.show()
```

```python
full_df = full_df.withColumn(
    'final_amount',
    col('total_amount') - col('discount')
)
full_df.show()

full_df.withColumnRenamed(
    'customer_name',
    'customer'
)
full_df.show()

full_df.withColumnRenamed(
    'order_status',
    'status'
)
full_df.show()

full_df.select(sum('total_amount').alias('total revenue')).show()
full_df.select(avg('total_amount').alias('avg order val')).show()
full_df.select(max('total_amount').alias('hig order val')).show()
full_df.select(min('total_amount').alias('low order val')).show()
print('order count: ', full_df.count())
df.groupby('city').count().show()
df.groupby('category').count().show()
```

```python
df.groupby('order_status').count().show()
df.groupby('city').sum('price').show()
df.groupby('category').sum('price').show()
df.groupby('product').sum('price').show()
df.groupBy('category').agg(avg('price').alias('avg_price')).show()
df.groupBy('category').agg(max('price').alias('max_price')).show()
df.groupBy('category').agg(min('price').alias('min_price')).show()
df.select('city').distinct().show()
df.select('category').distinct().show()
df.select('order_status').distinct().show()
```

```python
full_df.orderBy(full_df.final_amount.desc()).limit(3).show()
full_df.orderBy('final_amount').limit(1).show()
df.filter(df.product.like('M%')).show()
df.filter(df.customer_name.contains('a')).show()
df.filter(df.product.contains('o')).show()

from pyspark.sql.functions import when, col

new_df = full_df.withColumn(
    'order_value_category',
    when(col('total_amount') >= 50000, 'High')
    .when(col('total_amount') >= 20000, 'Medium')
    .otherwise('Low')
)
new_df.show()

new_df.groupBy('order_value_category').count().show()
full_df.filter(col('order_status') == 'Delivered').show()
full_df.filter(col('order_status') == 'Delivered').groupBy('city').agg(sum('final_amount').alias('delivered_order_revenue')).show()
```
